```python
from sklearn.utils import shuffle

df  = shuffle(crop_data,random_state=5)
df.head()
```

```python
x = df[['N', 'P','K','temperature', 'humidity', 'ph', 'rainfall']]
target = df['Crop']
```

```python
y = pd.get_dummies(target)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state= 0)

print("x_train :",x_train.shape)
print("x_test :",x_test.shape)
print("y_train :",y_train.shape)
print("y_test :",y_test.shape)
```

```python
from sklearn.datasets import make_classification
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

### Naive Bayes Classification

```python
gnb = GaussianNB()
model = MultiOutputClassifier(gnb, n_jobs=-1)
model.fit(x_train, y_train)
```

```python
gnb_pred = model.predict(x_test)
gnb_pred
```

```python
from sklearn.metrics import accuracy_score
a1 = accuracy_score(y_test.values.argmax(axis=1), gnb_pred.argmax(axis=1))
a1
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test.values.argmax(axis=1), gnb_pred.argmax(axis=1))
ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax);
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix');
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test.values.argmax(axis=1), gnb_pred.argmax(axis=1)))

print(metrics.classification_report(y_test.values.argmax(axis=1), gnb_pred.argmax(axis=1), digits=3))
```

### Random Forest Hyperparameter Tuning (GridSearchCV)

```python
from sklearn.model_selection import GridSearchCV
import time

param_grid = {
    'n_estimators': [50, 100, 150, 200],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

print("Starting RandomForest Hyperparameter Tuning...")
print("This may take a few minutes...\n")

rf_base = RandomForestClassifier(random_state=1)

grid_search = GridSearchCV(
    estimator=rf_base,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    scoring='accuracy',
    verbose=1
)

start_time = time.time()
grid_search.fit(x_train, y_train.values.argmax(axis=1))
end_time = time.time()

print(f"\nGrid Search completed in {(end_time - start_time):.2f} seconds")
print(f"\nBest Parameters: {grid_search.best_params_}")
print(f"Best Cross-Validation Score: {grid_search.best_score_:.4f}")
```

```python
best_rf_model = grid_search.best_estimator_

best_forest = MultiOutputClassifier(best_rf_model, n_jobs=-1)
best_forest.fit(x_train, y_train)

tuned_forest_pred = best_forest.predict(x_test)

from sklearn.metrics import accuracy_score
tuned_accuracy = accuracy_score(y_test.values.argmax(axis=1), tuned_forest_pred.argmax(axis=1))

print(f"Tuned RandomForest Accuracy: {tuned_accuracy:.4f}")
print(f"Default RandomForest Accuracy: {a3:.4f}")
print(f"Improvement: {(tuned_accuracy - a3)*100:.2f}%")
```

```python
from sklearn.metrics import confusion_matrix
cm_tuned = confusion_matrix(y_test.values.argmax(axis=1), tuned_forest_pred.argmax(axis=1))

ax = plt.subplot()
sns.heatmap(cm_tuned, annot=True, fmt='g', ax=ax)
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Tuned RandomForest - Confusion Matrix')
```

```python
from sklearn import metrics

print("="*50)
print("TUNED RANDOM FOREST - DETAILED METRICS")
print("="*50)
print("\nConfusion Matrix:")
print(metrics.confusion_matrix(y_test.values.argmax(axis=1), tuned_forest_pred.argmax(axis=1)))
print("\nClassification Report:")
print(metrics.classification_report(
    y_test.values.argmax(axis=1), 
    tuned_forest_pred.argmax(axis=1), 
    digits=3
))
```

```python
models = ['Naive Bayes', 'Decision Tree', 'RandomForest\n(Default)', 'RandomForest\n(Tuned)', 'KNN', 'Gradient Boost']
accuracies = [a1, a2, a3, tuned_accuracy, a4, a5]

plt.figure(figsize=(10, 6))
colors = ['#1f77b4', '#ff7f0e', '#d62728', '#2ca02c', '#9467bd', '#8c564b']
bars = plt.bar(models, accuracies, color=colors, alpha=0.8, edgecolor='black', linewidth=1.5)

for i, (model, accuracy) in enumerate(zip(models, accuracies)):
    plt.text(i, accuracy + 0.01, f'{accuracy:.4f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.ylabel('Accuracy', fontsize=12, fontweight='bold')
plt.title('Model Performance Comparison - All Classifiers', fontsize=14, fontweight='bold')
plt.ylim([0, 1.0])
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nSummary:")
print("-"*50)
for model, accuracy in zip(models, accuracies):
    print(f"{model:25s}: {accuracy:.4f}")
print("-"*50)
print(f"Best Model: {models[accuracies.index(max(accuracies))]} with accuracy {max(accuracies):.4f}")
```

```python
feature_names = x_train.columns
feature_importance = best_rf_model.feature_importances_

indices = np.argsort(feature_importance)[::-1]

plt.figure(figsize=(10, 6))
plt.title('Feature Importance - Tuned RandomForest Classifier', fontsize=14, fontweight='bold')
plt.bar(range(x_train.shape[1]), feature_importance[indices], align='center', alpha=0.8, color='#2ca02c', edgecolor='black')
plt.xticks(range(x_train.shape[1]), feature_names[indices], rotation=45, ha='right')
plt.ylabel('Importance Score', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nFeature Importance Ranking:")
print("-"*40)
for i in range(x_train.shape[1]):
    print(f"{feature_names[indices[i]]:15s}: {feature_importance[indices[i]]:.4f}")
```
